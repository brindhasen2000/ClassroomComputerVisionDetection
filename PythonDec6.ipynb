```python
a=(int(input("Enter weight:")))
b=(float(input("Enter height:")))
def bmi_calculator(a,b):
  bmi=a/b**2
  if bmi < 18.5:
    print("You're underweight. Eat more.","BMI: ",bmi)
  elif 18.5<=bmi<=24.9:
    print("You're normal.","BMI: ",bmi)
  elif 25<=bmi<=29.9:
    print("You're overweight, be more careful.","BMI: ",bmi)
  elif bmi>=30:
    print("Risk of obesity. Try a healthier diet.","BMI: ",bmi)
bmi_calculator(a,b)
```
