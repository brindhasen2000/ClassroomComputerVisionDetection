```python
# 31.GCD (Greatest Common Divisor)
print('To find GCD enter ')
num1 = int(input('number 1 : '))
num2 = int(input('number 2 : '))
if num1 > num2:
        greater = num2
else:
        greater = num1
for i in range(1, greater+1):
    if((num1 % i == 0) and (num2 % i == 0)):
        gcd = i
print('The gcd of' , num1 , 'and' , num2 , 'is' , gcd)
```

```python
# 32.LCM (Least Common Multiple)
print('To find LCM enter ')
num1 = int(input('number 1 : '))
num2 = int(input('number 2 : '))
if num1 > num2:
        lesser = num1
else:
        lesser = num2
while True:
    if(lesser % num1 == 0 and lesser % num2 == 0):
        lcm = lesser
        break
    lesser += 1

print('The lcm of' , num1 , 'and' , num2 , 'is' , lcm)
```

```python
# 33.to find the number of notes (Sample of notes: 10, 20, 50, 100, 500, and 1000 ) against an given amount
amount = int(input("Enter a number: "))
def num_of_notes(amount):
  count_1000 = 0
  count_500 = 0
  count_100 = 0
  count_50 = 0
  count_20 = 0
  count_10 = 0
  if amount>1000:
    count_1000 = amount//1000
    reminder = amount%1000
  if reminder > 500:
    count_500 = reminder//500
    reminder = reminder%500
  if reminder > 100:
    count_100 = reminder//100
    reminder = reminder%100
  if reminder > 50:
    count_50 = reminder//50
    reminder = reminder%50
  if reminder > 20:
    count_20 = reminder//20
    reminder = reminder%20
  if reminder >= 10:
    count_10 = reminder//10
  return count_1000, count_500, count_100, count_50, count_20, count_10
num_of_notes(amount)
```

```python
# 34.count vowels / consonants
word = input("Enter a word: ")
def count_vowel(word):
  vowels = ["a","e","i","o","u","A","E","I","O","U"]
  count_vowels = 0
  count_consonants = 0
  for alphabet in word:
    if alphabet in vowels:
      count_vowels = count_vowels + 1
    else:
      count_consonants = count_consonants + 1
  return count_vowels, count_consonants
count_vowel(word)
```
