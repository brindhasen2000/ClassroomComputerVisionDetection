```python

def balance_check(balance):
  return f"Your balance is {balance}"
def withdraw_amount(withdraw,balance):
  balance_amount = balance - withdraw
  return f"You have withdrew {withdraw} and your balance is {balance_amount}"
def deposit_amount(deposit,balance):
  new_balance = balance + deposit
  return f"You have deposited {deposit} and your balance is {new_balance}"
def new_pin_num(pin_num):
  return f"Your pin was {pin_num}. Your new pin is {new_pin}"


pin_num = 9576
user_pin_input = int(input("Enter pin: "))
balance = 10000
if user_pin_input == pin_num:
  options = "Enter your operations\nEnter 1 to check balance\nEnter 2 to withdraw\nEnter 3 to deposit\nEnter 4 to change pin\nEnter 5 to exit\n"
  user_input = input(f"{options}")

  if user_input == "1":
    print(balance_check(balance))
  if user_input == "2":
    withdraw = int(input("Enter withdrawl amount: "))
    print(withdraw_amount(withdraw,balance))
  if user_input == "3":
    deposit = int(input("Enter deposite amount: "))
    print(deposit_amount(deposit, balance))
  if user_input == "4":
    old_pin = int(input("Enter old pin: "))
    if old_pin == pin_num:
      new_pin = int(input("New pin: "))
      print("New pin set successfully.")
      pin_num = new_pin
    else:
      print("Old pin was wrong. Try with original pin.")
else:
  print("Wrong pin. Please try again.")
```
