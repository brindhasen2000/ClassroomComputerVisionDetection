```python
## multiple inheritance

class ClassC:
	def m1(self):
		print("In Class1")

class ClassA(ClassC):
	def m2(self):
		print("In Class2")

class ClassB(ClassC):
	def m3(self):
		print("In Class3")

obj_A = ClassA()
obj_A.m1()

obj_B = ClassB()
obj_B.m1()
```

```python
## hierarchical inheritance

class ClassB():
	def m2(self):
		print("In Class2")

class ClassC():
	def m3(self):
		print("In Class3")

class ClassA(ClassB, ClassC):
	def m4(self):
		print("In Class4")

obj_A = ClassA()
obj_A.m2()
obj_A.m3()
```

```python
##Abstraction

from abc import ABC, abstractmethod

# Define an abstract base class (ABC) representing a Shape
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
# Create concrete subclasses that inherit from Shape
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.1415 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.1415 * self.radius
class Rectangle(Shape):

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
# Create instances of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Use the abstracted methods without worrying about implementation details
print(f"Circle Area: {circle.area()}")
print(f"Circle Perimeter: {circle.perimeter()}")

print(f"Rectangle Area: {rectangle.area()}")
print(f"Rectangle Perimeter: {rectangle.perimeter()}")
```
