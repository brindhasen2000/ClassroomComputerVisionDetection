```python
import random
def game():
  user_input=input("Enter a choice(rock, paper, scissors): ")
  actions=["rock","paper","scissors"]
  comp_actions=random.choice(actions)
  print("You chose "+user_input+", computer chose "+comp_actions)
  if user_input==comp_actions:
    result = ("Tie")
  elif user_input=="rock" and comp_actions=="paper":
    result = ("Computer wins")
  elif user_input=="rock" and comp_actions=="scissors":
    result = ("User wins")
  elif user_input=="paper" and comp_actions=="rock":
    result = ("User wins")
  elif user_input=="paper" and comp_actions=="scissors":
    result = ("Computer wins")
  elif user_input=="scissors" and comp_actions=="rock":
    result = ("Computer wins")
  elif user_input=="scissors" and comp_actions=="paper":
    result = ("User wins")
  return result
a = int(input("How much?"))
computer = 0
user = 0
iteration = 1
# for i in range(a):
while iteration <= a:
  result = game()
  if result == "User wins":
    user = user+1
  if result == "Computer wins":
    computer=computer+1
  if result == "Tie":
    pass
  iteration = iteration + 1
if user > computer:
  print("User wins")
elif user < computer:
  print("Computer wins")
else:
  print("Tie")
```
