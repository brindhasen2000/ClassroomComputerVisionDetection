```python
##################### Regular Expression - REGEX #####################

# . - Selects all characters
# \d - digits(0-9)
# \D - except digits(0-9) everything else
# \w - matches (a-z, A-Z, 0-9, _)
# \W - matches everything except (a-z, A-Z, 0-9, _)
# \s - Whitespaces(space, tab, newline)
# \S - matches everything except Whitespaces(space, tab, newline)


# + - 1 or more characters
# * - 0 or more characters
# ? - 0 or one characters
# [] - Matches characters in brackets
# [^] - Matches characters
```

```python
import re
sample = 'Mr. Chow killed Mrs. Lane and the police started investigating'
pattern = r'M[rs|r|s].? [a-zA-Z]+'
matches = re.findall(pattern,sample)
print(matches)
```

```python
sample = "brindhasen2000@gmail.com"
gmail_pattern = r"[a-zA-Z0-9-$%!&]+@[a-z]+\.[a-z]+"
matches = re.findall(gmail_pattern,sample)
print(matches)
```

```python
gmail =  "darvin.joseph@gcitsolutions.com brindhasen2000@gmail.com these two emails are used for communication."
email_pattern = "[a-zA-Z0-9-$%!&.]+@[a-z]+\.[0-9a-z]+"  #### email Pattern

list_of_match = re.findall(email_pattern,gmail) #Findall matches everything and returns a list #Findall prioritizes groups before the actual match.
print(list_of_match)
```

```python
matches = re.finditer(email_pattern,gmail)
for i in matches:
  print(i.group())
```

```python
b = re.search(email_pattern,gmail)
print(b.group())
```

```python
c = re.compile(email_pattern)
print(c.sub("brindha",gmail))
```

```python
sample = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890


bing
ping
ming
cing
cycling


Ha HaHa


MetaCharacters (Need to be escaped):
.[{()\^$|?*+


augusto@yahoo.ca
bwcarty@att.net
an-dale@yahoo.com
mag123usnet@icloud.com
optimu$@att.net


321-555-4321
123.555.1234
421 879-5634


Mr. Todd Rudd
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
'''
```

```python
import re
pattern = r"[a-zA-Z]+ing"
matches = re.findall(pattern, sample)
print(matches)
```

```python
email_pattern = r"[a-zA-Z0-9!#$%&]+@[a-z]+\.[a-z0-9]+"
matches = re.findall(email_pattern, sample)
print(matches)
```

```python
phone_pattern = r"[0-9]+[-. ]+[0-9]+[-. ]+[0-9]+"
matches = re.findall(phone_pattern,sample)
print(matches)
```
