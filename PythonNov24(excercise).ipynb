```python
x=(input("Enter number:"))
input_list=list(x)
odd_product=1
even_sum=0
for i in input_list:
  if int(i)%2==0:
    print("this is even-"+str(i))
    even_sum+=int(i)
  else:
    print("this is odd-"+str(i))
    odd_product*=int(i)

#out of far loop, now printing the final values
print("This is even_sum : "+str(even_sum))
print("This is odd_product : "+str(odd_product))
```
